## Algorithm
- We first merge the blocks on the z-dimensional
- Two blocks can be merged if they are not overlapped in both dimension

```python
from collections import deque
import sys
sys.setrecursionlimit(10000)

input_path = "demo.txt"
data = []
with open(input_path, "r") as f:
    lines = f.readlines()
    for line in lines:
        line = line.replace("\n", "").strip()   
        data.append(list(line))

ROW, COL = len(data), len(data[0])
print(ROW, COL)
```

```python
import heapq
nodes = []
directions = [(0, 1), (0, -1), (1, 0), (-1, 0)]
distances = [[float("-inf") for _ in range(COL)] for _ in range(ROW)]
distance_queue = []
for i in range(ROW):
    for j in range(COL):
        if data[i][j] == "." and not nodes:
            heapq.heappush(nodes, (0, (i, j), set()))
            break
    
    if nodes:
        break

print(nodes)
while nodes:
    current_d, (current_r, current_c), stepped = heapq.heappop(nodes)
    if current_d <= distances[current_r][current_c]:
        continue
    print(current_r, current_c, stepped)

    distances[current_r][current_c] = current_d

    for dr, dc in directions:
        next_r, next_c = current_r + dr, current_c + dc
        if not (0 <= next_r < ROW and 0 <= next_c < COL):
            continue
        
        if data[next_r][next_c] == "#":
            continue
                
        if (next_r, next_c) in stepped:
            continue


        stepped.add((next_r, next_c))        
        heapq.heappush(nodes, (current_d + 1, (next_r, next_c), stepped.copy()))

distances[ROW-1][COL-2]
```
