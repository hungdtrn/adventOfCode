## Algorithm
- We first merge the blocks on the z-dimensional
- Two blocks can be merged if they are not overlapped in both dimension

```python
from collections import deque
import sys
sys.setrecursionlimit(10000)

input_path = "data.txt"
data = []
with open(input_path, "r") as f:
    lines = f.readlines()
    for line in lines:
        line = line.replace("\n", "").strip()   
        data.append(line)

ROW, COL = len(data), len(data[0])
print(ROW, COL)
```

```python
from collections import defaultdict
class Block:
    def __init__(self, row):
        left_range, right_range = row.split("~")
        left_range = [int(x) for x in left_range.split(",")]    
        right_range = [int(x) for x in right_range.split(",")]

        self.x_interval = left_range[0], right_range[0]
        self.y_interval = left_range[1], right_range[1]
        self.z_interval = left_range[2], right_range[2]

        self.next = []
        self.prev = []

    def __str__(self) -> str:
        return f"({self.x_interval}, {self.y_interval}, {self.z_interval}), prev: {len(self.prev)}, next: {len(self.next)}"
    
    def overlap(self, other):
        x_overlap, y_overlap = False, False
        if self.x_interval[0] <= other.x_interval[0] <= self.x_interval[1] or \
            self.x_interval[0] <= other.x_interval[1] <= self.x_interval[1] or \
            other.x_interval[0] <= self.x_interval[0] <= other.x_interval[1] or \
            other.x_interval[0] <= self.x_interval[1] <= other.x_interval[1]:
            x_overlap = True

        if self.y_interval[0] <= other.y_interval[0] <= self.y_interval[1] or \
            self.y_interval[0] <= other.y_interval[1] <= self.y_interval[1] or \
            other.y_interval[0] <= self.y_interval[0] <= other.y_interval[1] or \
            other.y_interval[0] <= self.y_interval[1] <= other.y_interval[1]:
            y_overlap = True
        
        return x_overlap & y_overlap


blocks = []
for row in range(ROW):
    blocks.append(Block(data[row]))

blocks = sorted(blocks, key=lambda x: x.z_interval[0])
# for b in blocks:
#     print(b)

stack = []
for b in blocks:
    remain = []
    while stack and not b.overlap(stack[-1]):
        remain.append(stack.pop())
    
    if stack:
        start_z = stack[-1].z_interval[1] 
        end_z = stack[-1].z_interval[1] 
    else:
        start_z = 0
        end_z = 0
    
    b.z_interval = (start_z + 1, end_z + b.z_interval[1] - b.z_interval[0] + 1)
    stack.append(b)
    stack.extend(remain)

for i in range(len(stack)):
    for j in range(i + 1, len(stack)):
        if stack[j].z_interval[1] == stack[i].z_interval[1]:
            continue
        
        if stack[j].z_interval[0] == stack[i].z_interval[1] + 1 and stack[j].overlap(stack[i]):
            stack[j].prev.append(stack[i])
            stack[i].next.append(stack[j])

ans = 0
for b in stack:
    can_removed = True
    for b1 in b.next:
        if len(b1.prev) <= 1:
            can_removed = False
            break
    print(b, can_removed)
    if can_removed:
        ans += 1
ans



```
